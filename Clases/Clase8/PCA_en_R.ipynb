```r
set.seed (848) 

sample <- datos %>%
  group_by(variedad)%>%
  sample_n(size=1000)
```

Antes de realizar el análisis de componentes principales se crea un dataframe con las variables numéricas (todas excepto la variable que clasifica según variedad de vino).

```r
df_numericas <- sample[,-13]
```

Se llevan a cabo los siguientes análisis:
*   PCA
*   PCA robusto: Determinante de mínima covarianza (MCD)
*   PCA robusto: Elipsoide de volumen mínimo (MVE)

Los métodos de PCA robusto son aplicados con el objetivo de evitar que los resultados del análisis se vean afectados por la presencia de outliers.

## PCA (no robusto)

```r
pca <- prcomp(df_numericas, scale = TRUE) # con datos estandarizados
summary(pca)
```

## PCA robusto: MCD (Determinante de mínima covarianza)

```r
pca_mcd <-princomp(df_numericas, cor=TRUE, scores=TRUE,
covmat=MASS::cov.mcd(df_numericas))#se especifica MCD
summary(pca_mcd)
```

## PCA robusto: MVE (Elipsoide de volumen mínimo)

```r
pca_mve <-princomp(df_numericas, cor=TRUE, scores=TRUE,
covmat=MASS::cov.mve(df_numericas))#se especifica MVE
summary(pca_mve)
```

## Comparación de resultados

La siguiente figura muestra el porcentaje de la varianza explicada por cada componente, para cada uno de los tres análisis realizados.

```r
install.packages("ggpubr")
library(ggpubr)
```

```r
par(mfrow=c(3,1))
p0 <-fviz_eig(pca, ncp =12, addlabels = TRUE, main="PCA convencional", barfill = "#c44367", barcolor = "#c44367")
p1 <-fviz_eig(pca_mcd, ncp =12, addlabels = TRUE, main="PCA - MCD", barfill = "#c44367", barcolor = "#c44367")
p2<- fviz_eig(pca_mve, ncp =12, addlabels = TRUE, main="PCA - MVE", barfill = "#c44367", barcolor = "#c44367")

ggarrange(p0,p1,p2, nrow = 1, ncol = 3)
```

```r
install.packages("data.table")              
library("data.table")
```

Para facilitar su interpretación se presenta la misma información de forma tabular.

```r
PoV_pca     <- pca$sdev^2/sum(pca$sdev^2)
PoV_pca_mcd <- pca_mcd$sdev^2/sum(pca_mcd$sdev^2)
PoV_pca_mve <- pca_mve$sdev^2/sum(pca_mve$sdev^2)

d_pca <- data.frame(as.list(PoV_pca))
d_pca_mcd <- data.frame(as.list(PoV_pca_mcd))
d_pca_mve <- data.frame(as.list(PoV_pca_mve))

setnames(d_pca, old = colnames(d_pca), new = colnames(d_pca_mcd))

proporciones <- rbind(d_pca, d_pca_mcd)
proporciones <- rbind(proporciones, d_pca_mve)
```

```r
p_tr = t(proporciones)
p_tr <- as.data.frame.matrix(p_tr) 
setnames(p_tr, old = colnames(p_tr), new = c('no_robusto', 'mcd', 'mve'))
p_tr
```

```r
p_tr <- p_tr %>% mutate(no_robusto_acum = cumsum(no_robusto)) %>% mutate(mcd_acum = cumsum(mcd)) %>% mutate(mve_acum = cumsum(mve))

p_tr[,c(4,5,6)]
```

La primera tabla muestra el porcentaje de varianza explicado por cada una de las componentes principales. La segunda presenta los mismos valores pero, esta vez, acumulados.

Salvo en el caso de que se requiera trabajar con una única componente principal, la mayor proporción de varianza explicada se alcanza con el análisis realizado con el estimador MVE.

## Conclusiones

Se aplicaron los métodos MCD (Determinante de mínima covarianza) y MVE (Elipsoide de volumen mínimo). Ambos lograron explicar, para una misma cantidad de dimensiones, una mayor proporción de la variabilidad de los datos respecto del PCA convencional. 
Estableciendo un umbral de porcentaje de varianza explicado de 80%, la dimensión del problema se reduce de 12 a 5. Si en cambio, ese umbral se fijara en 90% serían necesarias 7 componentes principales. 
Si bien el objetivo principal de esta metodología no es la clasificación de datos, los biplots asociados a PCA convencional y PCA-MVE muestran que puede obtenerse una clasificación preliminar de los vinos según su variedad a partir de solamente las dos primeras componentes principales. 
Observando las cargas resultantes de cada uno de los tres análisis, vemos que las variables originales ordenadas según su importancia de acuerdo a su relación con las distintas componentes principales, difieren según la técnica empleada (MCD o MVE). 
La graduación alcohólica y el contenido de ácido cítrico son dos de las variables de mayor importancia, es decir que explican gran parte de la variabilidad del conjunto de datos.
